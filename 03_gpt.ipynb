# Shakespeare GPT

```python
session2 = gpt2.start_tf_sess()
```

```python
gpt2.finetune(
    session2,
    "Shakespeare.txt",
    model_name="124M",
    steps=100,
    run_name="Shakespeare"
    )
```

```python
gpt2.generate(
    session2,
    prefix="Cycil: I Like Minors",
    run_name="shakespeare"
)
```
